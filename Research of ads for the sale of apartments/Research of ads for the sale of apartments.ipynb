```python
# колонка ceiling_height - высота потолков. Изучим статистику и гистограмму
print(data['ceiling_height'].describe())
data['ceiling_height'].hist(bins=12, range=(2, 5))
```

Strange minimum and maximum values, but we'll deal with anomalies later. The histogram shows that in most objects the ceiling height is from 2.5 meters to 3. Ceilings from 3 meters are less common. Before filling in the gaps, let's see if the ceiling height does not depend on other real estate parameters. At first glance, it seems that the ceiling height may depend on the city, the number of floors in the house, the floor, the number of rooms. Let's build the corresponding. charts

```python
# the ceiling height does not significantly depend on the city. There are outliers due to the small number of ads in the n.p.
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
print(data.pivot_table(index='locality_name', values='ceiling_height', aggfunc=['mean', 'count']))
```

```python
# the ceiling height depends on the number of floors in the house, but not very much
data.plot(x='floors_total', y='ceiling_height', kind='scatter', ylim=(2, 5), xlim=(0, 30), title='on the number of floors')

# there is a slight dependence on the floor of the property, similar to the previous one
data.plot(x='floor', y='ceiling_height', kind='scatter', ylim=(2, 5), title='on the floor of the property')

# there is no dependence at all on the number of rooms
data.plot(x='rooms', y='ceiling_height', kind='scatter', ylim=(2, 5), title='number of rooms')

# the picture is interesting from the distance from the center, but I will not use it to fill in the gaps, there are also gaps there
data.plot(x='cityCenters_nearest', y='ceiling_height', kind='scatter', ylim=(2, 5), title='distance from the center')
```

In houses with 8+ floors, the ceiling height changes significantly. Therefore, I will divide the data into two groups 0-7 and 8+ and in each I will fill in the gaps with the median value of the group

```python
# заполняю пропуски в высоте потолка

data.loc[(data['floor'] < 8) & (data['ceiling_height'].isna()), 'ceiling_height'] = \
data.loc[(data['floor'] < 8), 'ceiling_height'].median()

data.loc[(data['floor'] >= 8) & (data['ceiling_height'].isna()), 'ceiling_height'] = \
data.loc[(data['floor'] >= 8), 'ceiling_height'].median()

print('There are empty rows left in the column:', data['ceiling_height'].isna().sum())
```

#### Duplicates

We study implicit duplicates, then duplicate strings. Implicit duplicates can only be in the locality_name column. Using the loop, we will discard all auxiliary designations of cities. Out of 364, there are 305 unique values left. The drive to the lower case of unique values did not reduce. We can assume that duplicates in the names have been processed.

```python
# I collect unique values from a column, discarding nan and sort
locality_list = data['locality_name'].dropna().unique()
locality_list.sort()

not_use_types = (['городской', 'поселок', 'посёлок', 'деревня',
                'коттеджный', 'городского', 'типа', 'станции', 'при',
                'железнодорожной', 'садоводческое', 'некоммерческое',
                'товарищество', 'садовое', 'село'])

# Let's drop the types of n.p. and count the number of unique ones again
for i in range(len(data)):
    if type(data.at[i, 'locality_name']) == str:
        separated_name = data.at[i, 'locality_name'].split(' ')
        new_name_list = []
        for k in separated_name:
            if k not in not_use_types:
                new_name_list.append(k)
        new_name = ' '.join(new_name_list)
        data.at[i, 'locality_name'] = new_name


locality_list = data['locality_name'].dropna().unique()
locality_list.sort()
len(locality_list)
```

Objects with a ceiling height of up to 2.4 meters and more than 4.5 meters are excluded (21 and 38 respectively). The values were determined from a pivot table.

### Adding new columns to the table

the price of one square meter;
the day of the week of publication of the announcement (0 — Monday, 1 — Tuesday, and so on);
the month of publication of the announcement;
year of publication of the announcement;
type of apartment floor (values — "first", "last", "other");
distance to the city center in kilometers (convert from m to km and round to integer values).

```python
# m2_cost - cost per square meter
data['m2_cost'] = data['last_price'] / data['total_area']

# day_of_the_week - day of the week of publication
data['day_of_the_week'] = data['first_day_exposition'].dt.weekday

# month - month of publication of the ad
data['month'] = data['first_day_exposition'].dt.month

# year - year of publication
data['year'] = data['first_day_exposition'].dt.year

# floor_type - type of apartment floor

def floor_type_func(row):
    if row.loc['floor'] == 1:
        return 'первый'
    elif row.loc['floor'] == row.loc['floors_total']:
        return 'последний'
    else:
        return 'другой'

list_floor_type = data.apply(floor_type_func, axis=1)
data['floor_type'] = list_floor_type

# cityCenters_nearest_km - distance to the city center in kilometers
data['cityCenters_nearest_km'] = round(data['cityCenters_nearest'] / 1000)
```
